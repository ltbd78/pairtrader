```python
import requests
from bs4 import BeautifulSoup as bs
import yfinance as yf
from tqdm import tqdm
import pandas as pd
```

```python
# Get Ticker Names
url = 'https://topforeignstocks.com/stock-lists/the-complete-list-of-oil-gas-production-stocks-trading-on-the-nyse/'
r = requests.get(url)
soup = bs(r.text)
txts = []
for link in soup.find_all('a'):
    txts.append(link.text)

tickers = txts[12:113]
```

```python
# Download Data
data = []
for i in tqdm(range(len(tickers))):
    ticker = yf.Ticker(tickers[i])
    data.append(ticker.history(period='1y'))
```

```python
for d in data:
    print(len(d))
```

```python
# Clean Data
col_names = []
close = []
for i in range(len(data)):
    if len(data[i]) != 252:
        continue
    else:
        col_names.append(tickers[i])
        close.append(data[i]['Close'])
        
```

```python
df = pd.concat(close, axis=1, keys=col_names)
```
